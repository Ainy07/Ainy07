```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ['a' , 'b' , 'c' , 'd']
y = [10 , 25 , 18 , 30]

plt.bar(x , y)
plt.title("Simple Bar Chart")
plt.xlabel("category")
plt.ylabel("values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C', 'D']
sales_2023 = [20, 35, 30, 35]
sales_2024 = [25, 32, 34, 20]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, sales_2023, width, label='2023')
plt.bar(x + width/2, sales_2024, width, label='2024')

plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Grouped Bar Chart")
plt.xticks(x, labels)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
men = [20, 35, 30, 35]
women = [25, 32, 34, 20]

plt.bar(labels, men, label='Men')
plt.bar(labels, women, bottom=men, label='Women')

plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


sizes = [40,30,20,10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart with Percentage")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']

explode = [0.1, 0.2, 0, 0]
plt.pie(sizes, labels=labels, autopct="%1.1f%%" , explode=explode)
plt.title("Pie Chart with Explode")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']

colors = ['red', 'gold', 'purple', 'green']

plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors)
plt.title("Pie Chart with Custom Colors")
plt.show()
```

```python
import matplotlib.pyplot as plt
sizes = [40, 30, 20, 10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']
explode = [0.1, 0, 0, 0]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(6,6))
plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    explode=explode,
    shadow=True,
    startangle=50,
    colors=colors
)
plt.title("Beautiful Final Pie Chart")
plt.show()
```

```python
#Method 1: Simple Subplot (1 row, 2 columns)
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot([1,2,3,4],[10,20,25,30])
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(['A','B','C','D'],[5,7,3,8])
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
#Method 2: Grid Subplot (2 x 2 = total 4 graphs)
plt.figure(figsize=(10,8))

plt.subplot(2, 2, 1)
plt.plot([1,2,3], [3,2,5])
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.bar([1,2,3], [3,5,2])
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.scatter([1,2,3], [5,4,2])
plt.title("Plot 3")



plt.subplot(2, 2, 4)
plt.hist([10,20,20,30,40,50,50,60])
plt.title("Plot 4")

plt.tight_layout()
plt.show()
```

```python
#Method 3: Custom Layout with plt.subplots()

fig, ax = plt.subplots(1, 3, figsize=(12,4))

ax[0].plot([1,2,3],[2,4,1])
ax[0].set_title("Line")

ax[1].bar(['A','B','C'], [5,2,8])
ax[1].set_title("Bar")

ax[2].scatter([1,2,3],[3,6,1])
ax[2].set_title("Scatter")

plt.show()
```

```python
#Simple Pie Chart (base concept)

sizes = [40, 25, 20, 15]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Simple Pie Chart")
plt.show()
```

```python
#Donut Chart (Pie Chart + Hole)
import matplotlib.pyplot as plt

sizes = [40, 25, 20, 15]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
centre_circle = plt.Circle((0,0), 0.80, color='skyblue')
plt.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```

```python
#Pie Chart with Explode

sizes = [40, 25, 20, 15]
labels = ["A", "B", "C", "D"]
explode = [0.1, 0.3, 0, 0]

plt.pie(sizes, labels=labels, explode=explode, autopct="%1.1f%%")
plt.title("Pie Chart with Explode")
plt.show()
```

```python
#Double Ring Chart (Two-Level Donut)

fig, ax = plt.subplots(figsize=(6,6))

sizes1 = [40, 25, 20, 15]
ax.pie(sizes1, radius=0.9, labels=["A","B","C","D"], autopct="%1.1f%%")

sizes2 = [20, 30, 50]
ax.pie(sizes2, radius=0.67, labels=["X","Y","Z"], autopct="%1.1f%%")

centre = plt.Circle((0,0), 0.4, color="white")
ax.add_artist(centre)

plt.title("Double Ring Donut Chart")
plt.show()
```
