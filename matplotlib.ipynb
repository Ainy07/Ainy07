```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ['a' , 'b' , 'c' , 'd']
y = [10 , 25 , 18 , 30]

plt.bar(x , y)
plt.title("Simple Bar Chart")
plt.xlabel("category")
plt.ylabel("values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C', 'D']
sales_2023 = [20, 35, 30, 35]
sales_2024 = [25, 32, 34, 20]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, sales_2023, width, label='2023')
plt.bar(x + width/2, sales_2024, width, label='2024')

plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Grouped Bar Chart")
plt.xticks(x, labels)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
men = [20, 35, 30, 35]
women = [25, 32, 34, 20]

plt.bar(labels, men, label='Men')
plt.bar(labels, women, bottom=men, label='Women')

plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


sizes = [40,30,20,10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart with Percentage")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']

explode = [0.1, 0.2, 0, 0]
plt.pie(sizes, labels=labels, autopct="%1.1f%%" , explode=explode)
plt.title("Pie Chart with Explode")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']

colors = ['red', 'gold', 'purple', 'green']

plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors)
plt.title("Pie Chart with Custom Colors")
plt.show()
```

```python
import matplotlib.pyplot as plt
sizes = [40, 30, 20, 10]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']
explode = [0.1, 0, 0, 0]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(6,6))
plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    explode=explode,
    shadow=True,
    startangle=50,
    colors=colors
)
plt.title("Beautiful Final Pie Chart")
plt.show()
```
